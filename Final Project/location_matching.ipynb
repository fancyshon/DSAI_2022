```python
train_data=pd.read_csv('data/train.csv', usecols=['id','latitude', 'longitude', 'point_of_interest'], index_col=None)
train_data.head()
```

```python
# train_data.isnull().sum()
train, test= train_test_split(train_data, test_size = 0.2)
train_x, train_y=train.drop(['point_of_interest','id'],axis=1), train['point_of_interest']
test_x, test_y=test.drop(['point_of_interest','id'],axis=1), test["point_of_interest"]
# print(train_x)
# print(train_y)
```

```python
scaler = StandardScaler()
train_x=pd.DataFrame(scaler.fit_transform(train_x))
test_x=pd.DataFrame(scaler.fit_transform(test_x))
# print(len(test_x), len(test_y))
# print(len(train_x),len(train_y))
# print(len(test.drop('point_of_interest',axis=1)),len(test))

```

```python
loss=[]

for k in range(25):
    k+=1
    model=neighbors.KNeighborsClassifier(n_neighbors=k)
    model.fit(train_x, train_y)
    pred=model.predict(test_x)
    loss.append(np.mean(pred != test_y))
```

```python
plt.plot(loss, 'r')
plt.title('Error')
plt.legend()
plt.show()
```
